```python
import numpy as np
import pandas as pd
import csv
import time
import random
random.seed()

df = pd.read_json('dataset/processed_restaurant_reviews_1.json')
data = df.text.apply(' '.join)[:300000]
del(df)


```

```python
sizes = [10, 15, 20, 25, 30, 40]
i = 2 # Change this to open labels for different cluster size indexing using sizes array
thefile = open('dataset/300ktop_category_' + str(sizes[i]) + '.txt', 'r')
reader = csv.reader(thefile)
labels = [] # This will store the labels
for row in reader:
    labels.append(row[0])
thefile.close()
labels = np.array(labels)
curr_label = 3 #change this to ouput reviews from Topic #(curr_label)

cap = len(data[labels==str(curr_label)])
if cap<100:
    thing = random.sample(data[labels==str(curr_label)],cap)
else:
    thing = random.sample(data[labels==str(curr_label)],100)
for j in thing:
    print j, '\n'
```

```python
allData = []
for i in range(6):
    thefile = open('dataset/300ktop_category_' + str(sizes[i]) + '.txt', 'r')
    reader = csv.reader(thefile)
    labels = np.zeros(40) # This will store the labels
    for row in reader:
        labels[int(row[0])]+=1
    thefile.close()
    allData.append(labels.copy())
```

```python
df = pd.DataFrame(allData,index = sizes).T
```
